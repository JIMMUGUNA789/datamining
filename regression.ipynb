```python
import pandas as pandas
import matplotlib.pyplot as pyplot
import numpy as numpy
import scipy.stats as stats
import seaborn as seaborn
from matplotlib import rcParams
# Ensure that all plots show up in your notebook
%matplotlib inline
%pylab inline
# read the csv file
dataFile = pandas.read_csv('./kcHouseData.csv')
# use the first line in the csv file as th column titles .head() takes an int argument with 0 being the default
dataFile.head()
```

```python
# check if there are any null rows
dataFile.isnull().any()
```

```python
# check the data types of the columns
dataFile.dtypes
```

```python
# Statistical summary of the data
dataFile.describe()
```

```python
# draw a histogram
fig = pyplot.figure(figsize=(12,6))
squareFeet = fig.add_subplot(121)
cost = fig.add_subplot(122)
squareFeet.hist(dataFile.sqft_living, bins=80)
squareFeet.set_xlabel('FT^2')
squareFeet.set_title("Histogram of House Square Footage")

cost.hist(dataFile.price, bins=80)
cost.set_xlabel('Price($)')
cost.set_title("Histogram of Housing Prices")

pyplot.show()
```

```python
import statsmodels.api as statsmodels
# import Ordiary Least Squares
# formulae:
#Reg = ols('Dependent variable-independent variable(s), dataframe').fit()
# print(Reg.summary())
from statsmodels.formula.api import ols
m = ols('price ~ sqft_living', dataFile).fit()
print(m.summary())
```

```python
m = ols('price ~ sqft_living + bedrooms + grade + condition',dataFile).fit()
print (m.summary())
```

```python
seaborn.jointplot(x="sqft_living", y="price", data=dataFile, kind = 'reg',fit_reg= True)
pyplot.show()
```
